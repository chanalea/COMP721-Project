```
games = games.loc[games['SEASON'] == 2004]

games = games.loc[games["HOME_TEAM_ID"] != 1610612766]
games = games.loc[games["VISITOR_TEAM_ID"] != 1610612766]
games
```

```
team_season = team_season.loc[team_season['year'] == 2004]
team_season = team_season.drop(columns= ['leag'], axis=1)
team_season
```

```
col_list =["HOME_TEAM_ID","VISITOR_TEAM_ID","SEASON", "HOME_TEAM_WINS"]
game_outcome = games[['HOME_TEAM_ID', 'VISITOR_TEAM_ID','HOME_TEAM_WINS']]
game_outcome = np.array(game_outcome)
game_outcome
```

```
team_ID = [1610612737, 1610612738, 1610612741, 1610612739, 1610612742, 1610612743, 1610612765, 1610612744, 1610612745, 1610612754, 1610612746, 1610612747, 1610612763, 1610612748, 1610612749, 1610612750, 1610612751, 1610612740, 1610612752, 1610612753, 1610612755, 1610612756, 1610612757, 1610612758, 1610612759, 1610612760, 1610612761, 1610612762, 1610612764]
```

```
team_seas_2004 = np.array(team_season)
team_stats = []
for r in range(len(team_seas_2004)):
  x = []
  x.append(r)
  for c in range(2,team_seas_2004.shape[1]):
    x.append(team_seas_2004[r][c])

  team_stats.append(x)

print(len(team_stats[0]))
team_stats
```

# Feature Vector

```
data = []
for i in range(len(game_outcome)):
  # home team stats
  find_index = np.where(team_ID == game_outcome[i][0])
  index = find_index[0][0]
  home_side = team_stats[index]
 
  # away team stats
  find_index = np.where(team_ID == game_outcome[i][1])
  index = find_index[0][0]
  away_side =  team_stats[index]

   # 0 if home team won or 1 if home team lost 
  outcome = []
  outcome.append(game_outcome[i][2])
  
  # add feature vector made from hometeam stats, awayteam stats, outcome of game
  data.append(np.concatenate((home_side, away_side,outcome), axis=None))
```

```
print(len(data[0]))
print(data)
```

```
data = np.array(data)
print(data)
```

#Calculations and stuff

```
from sklearn import metrics
def analysis(ypred, ytest):
  print ("Accuracy Score: ", metrics.accuracy_score(ytest, y_pred))
  print("Average Precision: ", metrics.average_precision_score(ytest, y_pred))
  print("Balanced Accuracy", metrics.balanced_accuracy_score(ytest, y_pred))
  print("Balanced F1 score", metrics.f1_score(ytest, y_pred))
  print("Average Hamming Loss", metrics.hamming_loss(ytest, y_pred))
  print("Log loss", metrics.log_loss(ytest, y_pred))
  print("Precision Score", metrics.precision_score(ytest, y_pred))
  print("Recall Score", metrics.recall_score(ytest, y_pred))
  print("Confusion Matrix \n", metrics.confusion_matrix(ytest, y_pred))
  print("ROC Curve", metrics.roc_curve(ytest, y_pred))
  
```

# Classification (using data as is)

```
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
```

```
train, test = train_test_split( data, test_size = 0.1,random_state = 20)

xtrain = train[:, :-1] # for all but last column
ytrain = train[:, -1] # for last column (result label )

xtest = test[:, :-1] 
ytest = test[:, -1] 

print("Training set size: %.0f" % len(xtrain))
print("Testing set size: %.0f" % len(xtest))
```

```
clf = svm.SVC(kernel = 'linear') # Linear Kernel

clf.fit(xtrain, ytrain)

y_pred = clf.predict(xtest)
print("Support Vector Classifier")
analysis(y_pred, ytest)
metrics.plot_roc_curve(clf, xtest, ytest)
plt.show()
```

```
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr', max_iter=500)
clf.fit(xtrain, ytrain)
print("LogisticRegression")
y_pred = clf.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(clf, xtest, ytest)
plt.show()
```

```
clf = SGDClassifier(loss= 'modified_huber', shuffle=True, random_state=1)
clf.fit(xtrain, ytrain)
print("SGD Classifier")
y_pred = clf.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(clf, xtest, ytest)
plt.show()
```

```
clf = MLPClassifier(alpha=1e-5, hidden_layer_sizes=(100, 10), random_state=1)
clf.fit(xtrain, ytrain)
print("MLPClassifier")
y_pred = clf.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(clf, xtest, ytest)
plt.show()
```

```
rf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
rf.fit(xtrain, ytrain)
print("Random Forest")
y_pred = rf.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(rf, xtest, ytest)
plt.show()
```

```
abc = AdaBoostClassifier(n_estimators=50,learning_rate=0.2)
abc.fit(xtrain, ytrain)
print("AdaBoost Classifier")
y_pred = abc.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(abc, xtest, ytest)
plt.show()
```

```
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(xtrain, ytrain)
print("KNClassifier")
y_pred = kn.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(kn, xtest, ytest)
plt.show()
```

```
pa = PassiveAggressiveClassifier()
pa.fit(xtrain, ytrain)
print("Passive Aggresive")
y_pred = pa.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(pa, xtest, ytest)
plt.show()
```

```
bagging = BaggingClassifier(base_estimator=PassiveAggressiveClassifier(),
                            n_estimators=10, random_state=0)
bagging.fit(xtrain, ytrain)
print("Bagging Classifier")
y_pred = bagging.predict(xtest)
analysis(y_pred, ytest)
metrics.plot_roc_curve(bagging, xtest, ytest)
plt.show()
```
